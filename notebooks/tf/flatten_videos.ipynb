```python
from glob import glob

paths = glob("../../persist/blob/channels/*/video_list/*")

paths
```

```python
import json

vid_data = []
for file in paths:
    with open(file) as js_io:
        vid_dict = json.load(js_io)
    vid_info = [x["snippet"] | x["contentDetails"] for x in vid_dict["items"]]
    vid_data.extend(vid_info)
```

```python
import pandas as pd
vid_df = pd.DataFrame(vid_data)
vid_df
```

```python
climate_words = [
    "climate",
    "global warming",
    "ice caps",
    "biodegradable",
    "carbon dioxide",
    "carbon monoxide",
    "coal",
    "oil",
    "petrol",
    "emission",
    "green energy",
    "environmental",
    "fossil fuel",
]

climate_str = '|'.join(climate_words)

climate_str
```

```python
environmental_df = vid_df[
    vid_df["title"].str.contains(climate_str) |
    vid_df["description"].str.contains(climate_str)
]

environmental_df.to_csv("../../persist/cleansed/environmental_videos.csv", index=False)
```
